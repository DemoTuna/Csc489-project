```python
import os
import glob
import cv2
import numpy as np
from ultralytics import YOLO

model = YOLO("runs/detect/train/weights/best.pt")

base_path = "/content/drive/MyDrive/arabic_sign_language.v1-version_1.yolov12"
test_images = glob.glob(f"{base_path}/test/images/*.jpg")

class_names = model.names
num_classes = len(class_names)

ious_per_class = [[] for _ in range(num_classes)]

def compute_iou(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    inter = max(0, xB - xA) * max(0, yB - yA)

    areaA = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    areaB = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])
    if areaA + areaB - inter == 0:
        return 0
    return inter / (areaA + areaB - inter)

def yolo_to_xyxy(x, y, w, h, img_w, img_h):
    x1 = (x - w/2) * img_w
    y1 = (y - h/2) * img_h
    x2 = (x + w/2) * img_w
    y2 = (y + h/2) * img_h
    return [x1, y1, x2, y2]

for img_path in test_images:
    img = cv2.imread(img_path)
    h, w = img.shape[:2]

    pred = model(img_path, verbose=False)[0]

    # LOAD GT LABELS
    label_path = img_path.replace("/images/", "/labels/").replace(".jpg", ".txt")

    if not os.path.exists(label_path):
        continue

    with open(label_path, 'r') as f:
        gt_boxes = []
        for line in f:
            cls, xc, yc, bw, bh = map(float, line.split())
            gt_xyxy = yolo_to_xyxy(xc, yc, bw, bh, w, h)
            gt_boxes.append((int(cls), gt_xyxy))

    # MATCH PREDICTIONS
    for box in pred.boxes:
        pred_cls = int(box.cls)
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        pred_box = [x1, y1, x2, y2]

        # FIND GT of same class
        for (gt_cls, gt_box) in gt_boxes:
            if gt_cls == pred_cls:
                iou = compute_iou(pred_box, gt_box)
                ious_per_class[pred_cls].append(iou)

# PRINT RESULTS
print("\n=== IoU PER CATEGORY ===")
for cls_id in range(num_classes):
    if len(ious_per_class[cls_id]) > 0:
        print(f"{cls_id} ({class_names[cls_id]}): {np.mean(ious_per_class[cls_id]):.4f}")
    else:
        print(f"{cls_id} ({class_names[cls_id]}): No matches")

print("\n=== OVERALL IoU ===")
all_iou = [iou for cls_list in ious_per_class for iou in cls_list]
print(f"{np.mean(all_iou):.4f}")
```
